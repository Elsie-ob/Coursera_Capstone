## PART1 - EXPLORING AND CLUSTERING NEIGHBOURHOODS IN CANADA

```python
wiki = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
wikipedia_page = requests.get(wiki)
```

```python
df_one = pd.read_html(wikipedia_page.content, header=0)[0]
df_new = df_one[df_one.Borough != 'Not assigned']

df_new.head()
```
